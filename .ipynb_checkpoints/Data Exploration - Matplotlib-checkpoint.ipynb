```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
red_wine_df = pd.read_csv("Resources/winequality-red.csv", sep=';')
white_wine_df = pd.read_csv("Resources/winequality-white.csv", sep=';')
baseline_wine_df = pd.concat([red_wine_df, white_wine_df])
```

```python
#white wine values are only from 3 to 9
white_count = white_wine_df.groupby("quality").count()
white_count
```

```python
new_white_count = white_count.copy()
new_white_count.reset_index(inplace=True)
w_count = new_white_count[["quality","fixed acidity"]]
w_count.rename(columns={"fixed acidity":"Total white wine count"})
```

```python
#red wine values are only from 3 to 8
red_count = red_wine_df.groupby("quality").count()
red_count
```

```python
new_red_count = red_count.copy()
new_red_count.reset_index(inplace=True)
r_count = new_red_count[["quality","fixed acidity"]]
r_count.rename(columns={"fixed acidity":"Total red wine count"})
```

```python
#Groupby quality but means for each value
gb_qual = white_wine_df.groupby("quality")
gb_qual.mean()
```

```python
#Groupby quality but means for each value
gb_qual = red_wine_df.groupby("quality")
gb_qual.mean()
```

```python
bin_values = np.arange(start=0, stop=10, step=2)
red_wine_df['quality'].hist(bins=bin_values, figsize=[14,6])
plt.title("1599 Red Wines")
plt.ylabel("Number of Wines")
plt.xlabel("Quality scores")
plt.show()
```

```python
bin_values = np.arange(start=0, stop=10, step=1)
white_wine_df['quality'].hist(bins=bin_values, figsize=[14,6])
plt.title("4898 White Wines")
plt.ylabel("Number of Wines")
plt.xlabel("Quality scores")
plt.show()
```

```python
plt.figure(figsize=(15,15))
matrix = np.tril(baseline_wine_df.corr())
sns.heatmap(baseline_wine_df.corr(), annot=True, mask=matrix,cmap="coolwarm")
plt.show()
```

```python
#Both white and red wine combined data
plt.figure(figsize=(15,15))
sns.heatmap(baseline_wine_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
#Both white wine combined data
plt.figure(figsize=(15,15))
sns.heatmap(white_wine_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
#Both red wine combined data
plt.figure(figsize=(15,15))
sns.heatmap(red_wine_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```
